```python
import pandas as pd
import numpy as np
import glob
```

```python
files = glob.glob("data/novamart_orders_2024_*.csv")

df = pd.concat(
    [pd.read_csv(file) for file in files],
    ignore_index=True
)
```

```python
df = df.drop_duplicates()
```

```python
df["sales"] = (
    df["sales"]
      .str.replace("$","",regex=False)
      .str.replace(",","",regex=False)
      .astype(float)
)
```

```python
df["order_date"] = pd.to_datetime(
    df["order_date"],
    format="mixed"
)

df["ship_date"] = pd.to_datetime(
    df["ship_date"],
    format="mixed"
)
```

```python
df["category"] = df["category"].str.strip().str.title()

df["region"] = df["region"].str.strip().str.title()
```

```python
df["discount"] = df["discount"].fillna(0)

df["profit"] = df["profit"].fillna(0)
```

```python
df["order_month"] = df["order_date"].dt.month_name()
```

```python
df["profit_margin_pct"] = (
    df["profit"] /
    df["sales"]
)*100
```

```python
df["shipping_days"] = (
    df["ship_date"] -
    df["order_date"]
).dt.days
```

```python
product = pd.read_csv("data/product_lookup.csv")

df = df.merge(
    product,
    on="product_id",
    how="left"
)
```

```python
df.to_csv(
    "data/novamart_clean.csv",
    index=False
)
```
